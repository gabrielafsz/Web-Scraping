## Web Scraping

**Objetivo:** Extrair ficha técnica de produtos a venda em algum site de compra.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Usaremos a biblioteca requests para extrair os dados do site ``requests.get()``.

Caso seja necessário, usar headers e timeout para evitar problemas de acesso ou excesso de requisição.

Para obter o user-agent, basta acessar o browser do computador e digitar: my user agent.

O timeout é o tempo limite para realizar a solicitação.

```python
url = 'https://www.taqi.com.br/fogoes/fogao/cat263630164'

cabecalho = { 'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/109.0'}
response = requests.get(url, headers= cabecalho)
response
```

Response [200] significa que deu certo! Agora você pode acessar os dados do site usando a biblioteca BeautifulSoup, que organiza e cria uma estrutura hierárquica.

```python
site = BeautifulSoup(response.text, 'html.parser')
print(site.prettify())
```

Os próximos passos são:

    - acessar o site que deseja extrair dados
    - abrir o console usando o F12
    - selecionar na página o elemento que você necessita 
    
O elemento selecionado corresponde a um script em html no console. É por meio desse script que você se orienta no momento da busca em ``.find_all()``.

```python
soup = site
list_link = []
for link in soup.find_all('a'): 
    list_link.append(link.get('href'))
```

### Queimadores

```python
#url = 'https://www.taqi.com.br/fogao-embutir-dako-glass-5-bocas/211267'
url = 'https://www.zoom.com.br/lead?oid=368725216&channel=11&index=7&sortorder=7&pagesize=13&appliedpromotion=CASHBACK'

cabecalho = { 'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/109.0'}
response = requests.get(url, headers= cabecalho, timeout=10)
response
```

```python
site = BeautifulSoup(response.text, 'html.parser')
print(site.prettify())
```

```python
texto = str(site.find_all('div', class_="DetailsSection_MerchantContent__45_Z4")[0]).split('<br/>')
texto
```

```python
df = pd.DataFrame(columns=['info'])
df
```

```python
itens=[]
for i in range(0,len(texto),1):
    itens.append(texto[i])

df['info'] = itens
```

```python
import numpy as np

df['Col1'] = (np.where(df['info'].str.contains(':'),
                  df['info'].str.split(':').str[0],
                  df['info']))

df['Col2'] = (np.where(df['info'].str.contains(':'),
                  df['info'].str.split(':').str[1],
                  df['info']))
```

```python
df.set_index('Col1', inplace=True)
```

```python
col, val =[],[]
i = 0
for item in site.find_all('td'):
    i = i+1
    item = item.text
    if i%2 :
        col.append(item)
    else:
        val.append(item)
        
df = pd.DataFrame(columns=['nome','valor'])
df['nome'] = col
df['valor'] = val
```
